```python
import pandas as pd
import numpy as np
```

```python
###Loading the features of test data set
data2=pd.read_csv("test_set_features.csv")
```

```python
###Loading the features of train data set
data1=pd.read_csv("training_set_features.csv")
data1.describe()
```

```python
### Loading target values of training dataset
out=pd.read_csv("training_set_labels.csv")
out1=out[['xyz_vaccine']]
out2=out[['seasonal_vaccine']]
out.describe()
```

```python
### Converting categorial data into numerical data for training dataset
data1['age_group'].replace(['55 - 64 Years', '35 - 44 Years', '18 - 34 Years', '65+ Years',
       '45 - 54 Years'],
                        [4,2,1,5,3], inplace=True)
data1['education'].replace(['< 12 Years', '12 Years', 'College Graduate', 'Some College'],
                        [1,2,4,3], inplace=True)
data1['race'].replace(['White', 'Black', 'Other or Multiple', 'Hispanic'],
                        [1,2,3,4], inplace=True)
data1['sex'].replace(['Female', 'Male'],
                        [0,1], inplace=True)
data1['income_poverty'].replace(['Below Poverty', '<= $75,000, Above Poverty', '> $75,000'],
                        [0,1,2], inplace=True)
data1['marital_status'].replace(['Not Married', 'Married'],
                        [0,1], inplace=True)
data1['rent_or_own'].replace(['Own', 'Rent'],
                        [1,0], inplace=True)
data1['employment_status'].replace(['Not in Labor Force', 'Employed', 'Unemployed'],
                        [1,2,0], inplace=True)
hhs_map=data1['hhs_geo_region'].value_counts().to_dict()
data1['hhs_geo_region']=data1['hhs_geo_region'].map(hhs_map)
data1['census_msa'].replace(['Non-MSA', 'MSA, Not Principle  City', 'MSA, Principle City'],
                        [0,1,2], inplace=True)
ei_map=data1['employment_industry'].value_counts().to_dict()
data1['employment_industry']=data1['employment_industry'].map(ei_map)
eo_map=data1['employment_occupation'].value_counts().to_dict()
data1['employment_occupation']=data1['employment_occupation'].map(eo_map)
```

```python
### Converting categorial data into numerical data for test dataset
data2['age_group'].replace(['55 - 64 Years', '35 - 44 Years', '18 - 34 Years', '65+ Years',
       '45 - 54 Years'],
                        [4,2,1,5,3], inplace=True)
data2['education'].replace(['< 12 Years', '12 Years', 'College Graduate', 'Some College'],
                        [1,2,4,3], inplace=True)
data2['race'].replace(['White', 'Black', 'Other or Multiple', 'Hispanic'],
                        [1,2,3,4], inplace=True)
data2['sex'].replace(['Female', 'Male'],
                        [0,1], inplace=True)
data2['income_poverty'].replace(['Below Poverty', '<= $75,000, Above Poverty', '> $75,000'],
                        [0,1,2], inplace=True)
data2['marital_status'].replace(['Not Married', 'Married'],
                        [0,1], inplace=True)
data2['rent_or_own'].replace(['Own', 'Rent'],
                        [1,0], inplace=True)
data2['employment_status'].replace(['Not in Labor Force', 'Employed', 'Unemployed'],
                        [1,2,0], inplace=True)
hhs_map2=data2['hhs_geo_region'].value_counts().to_dict()
data2['hhs_geo_region']=data2['hhs_geo_region'].map(hhs_map2)
data2['census_msa'].replace(['Non-MSA', 'MSA, Not Principle  City', 'MSA, Principle City'],
                        [0,1,2], inplace=True)
ei_map=data2['employment_industry'].value_counts().to_dict()
data2['employment_industry']=data2['employment_industry'].map(ei_map)
eo_map=data2['employment_occupation'].value_counts().to_dict()
data2['employment_occupation']=data2['employment_occupation'].map(eo_map)
```

```python
### Filling Nan values with mode 
from sklearn.impute import SimpleImputer 
for i in range(36):
    X = data1.iloc[:, i].values.reshape(-1, 1) 
      
    # To calculate mean use imputer class 
    imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent') 
    imputer = imputer.fit(X) 
      
    X = imputer.transform(X) 
    data1.iloc[:, i]=X
```

```python
from sklearn.impute import SimpleImputer 
for i in range(36):
    X = data2.iloc[:, i].values.reshape(-1, 1) 
      
    # To calculate mean use imputer class 
    imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent') 
    imputer = imputer.fit(X) 
      
    X = imputer.transform(X) 
    data2.iloc[:, i]=X
```
